1.	What is Regression?

```python
Regression is a statistical and machine learning technique used to model and analyze the relationship between a dependent variable and one or more independent variables.
It is mainly used to predict continuous values such as price, salary, or temperature.
```

2.	What is Linear Regression?

```python
Linear Regression is a type of regression that models the relationship between variables using a straight line.
It predicts the dependent variable by fitting a linear equation to the observed data.
```

3.	When to use Linear Regression? Explain the equation of a straight line.

```python
Linear Regression is used when there is a linear relationship between independent and dependent variables and we want to predict a continuous value.
The equation of a straight line is:
y = mx + b

Where:
y = dependent variable
x = independent variable
m = slope of the line
b = intercept (value of y when x = 0)
```

4.	What kind of plots will you use to showcase the relationship amongst the columns?

```python
To show relationships between columns we use:

Scatter Plot – shows relationship between two variables
Pair Plot – shows relationships between multiple variables
Heatmap (correlation matrix) – shows correlation between columns
```

5.	How is the best fit line chosen?

```python
The best fit line is chosen by minimizing the error between actual values and predicted values.
This is done using the Least Squares Method, which minimizes the sum of squared differences (errors) between predicted and actual values.
```

6.	What is gradient descent, and why is it used? Explain the maths.

```python
Gradient Descent is an optimization algorithm used to minimize the cost function in machine learning models.

It updates model parameters using:
θ = θ − α ∂J(θ)/∂θ
Where θ is the parameter, α is the learning rate, and J(θ) is the cost function.
It repeatedly updates parameters until the prediction error becomes minimum.
```

7.	What are residuals?

```python
Residuals are the difference between the actual value and the predicted value in a regression model.
Residual = Actual value − Predicted value
```

8.	What is correlation?

```python
Correlation measures the strength and direction of the relationship between two variables.
Its value ranges from -1 to +1.
```

9.	What is multicollinearity?

```python
Multicollinearity occurs when two or more independent variables in a regression model are highly correlated with each other, making it difficult to determine their individual effect on the dependent variable.
```

10.	How to detect multicollinearity?

```python
Multicollinearity can be detected using:

1) Correlation matrix / heatmap
2) Variance Inflation Factor (VIF)
3) Scatter plots between variables
```

11.	What are the remedies for multicollinearity?

```python
Remedies include:

1) Removing highly correlated variables
2) Combining correlated features
3) Using techniques like Ridge or Lasso regression
```

12.	What is the R-Squared Statistics?

```python
R-squared (R²) measures how well the regression model explains the variability of the dependent variable.
Its value ranges from 0 to 1, where higher values indicate a better fit.
```

13.	What is an adjusted R-Squared Statistics?

```python
Adjusted R² is a modified version of R² that adjusts for the number of independent variables in the model.
It provides a more accurate measure of model performance when multiple predictors are used.
```

14.	Why do we use adj R-squared?

```python
Adjusted R² is used because it considers the number of independent variables in the model and gives a more accurate measure of model performance than R².
```

15.	Why adj R-squared decreases when we use incompetent variables?



```python
Adjusted R² decreases when irrelevant variables are added because they do not improve the model’s predictive power, and adjusted R² penalizes unnecessary variables.
```

16.	How to interpret a Linear Regression model?

```python
A Linear Regression model is interpreted by examining the coefficients, which show how much the dependent variable changes for a one-unit change in an independent variable, keeping other variables constant.
```

17.	What is the difference between fit, fit_transform and predict methods?

```python
fit() – Trains the model using the training data.
fit_transform() – Fits the model and transforms the data in one step.
predict() – Uses the trained model to predict output values.
```

18.	How do you plot the least squared line?

```python
First create a scatter plot of the data points, then calculate predictions using the regression model and plot the regression line over the scatter plot.
```

19.	What are Bias and Variance? What is Bias Variance Trade-off?

```python
Bias is the error due to overly simple assumptions in the model.
Variance is the error due to the model being too sensitive to training data.
Bias-Variance Trade-off is balancing both to achieve the best model performance.
```

20.	What is the null and alternate hypothesis?

```python
Null Hypothesis (H₀): Assumes no relationship or no effect between variables.
Alternative Hypothesis (H₁): Assumes there is a relationship or effect between variables.
```

21.	What is multiple linear regression?

```python
Multiple Linear Regression is an extension of linear regression where two or more independent variables are used to predict a dependent variable.
```

22.	What is the OLS method? Derive the formulae used in the OLS method.

```python
OLS (Ordinary Least Squares) is a method used to estimate regression coefficients by minimizing the sum of squared residuals between actual and predicted values.

The objective function minimized is:
Σ (yᵢ − ŷᵢ)²
Where yᵢ is the actual value and ŷᵢ is the predicted value.
```

23.	What is the p-value? How does it help in feature selection?

```python
The p-value measures the statistical significance of a variable.
In feature selection, variables with p-value less than a threshold (commonly 0.05) are considered important, while variables with higher p-values may be removed.
```

24.	How to handle categorical values in the data?

25.	What is regularization, and why do we need it?

```python
Regularization is a technique used to reduce overfitting by adding a penalty term to the loss function, which helps keep model coefficients small and improves generalization.
```

26.	Explain Ridge Regression.

```python
Ridge Regression is a type of regularized linear regression that adds an L2 penalty (squared coefficients) to the loss function to reduce overfitting and multicollinearity.
```

27.	Explain Lasso Regression.

```python
Lasso Regression is another regularization method that adds an L1 penalty (absolute value of coefficients) to the loss function, which can shrink some coefficients to zero and perform feature selection.
```

28.	Explain Elastic Net.

```python
Elastic Net is a regularization technique that combines both Ridge (L2) and Lasso (L1) penalties. It helps reduce overfitting and also performs feature selection.
```

29.	Why do we do a train test split?

```python
Train-test split is done to evaluate the performance of a machine learning model on unseen data.
The model is trained on the training set and tested on the test set.
```

30.	What is polynomial regression? When to use it?

```python
Polynomial Regression is a type of regression where the relationship between variables is modeled as a polynomial (curved line) instead of a straight line.
It is used when the relationship between variables is non-linear.
```
